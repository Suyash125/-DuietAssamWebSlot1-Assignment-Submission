```python
price1=float(input("Enter the amount of the first item: $"))
```

```python
price2=float(input("Enter the amount of the second item: $"))
```

```python
subtotal=price1+price2
```

```python
tip=subtotal*.10
```

```python
print("\nSubtotal: $", format(subtotal, '.2f'), sep="")
```

```python
print('A 10%% tip is: $%.2f.' %(tip))
```

```python
print('Total price with 10%% tip is: $%.2f' % (subtotal+tip))
```
